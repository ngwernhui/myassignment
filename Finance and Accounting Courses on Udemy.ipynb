## 1. Exploring the dataset

```python
#data manipulation
import pandas as pd
import numpy as np

#data visualization
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [10, 6]
```

```python
finance_udemy = pd.read_csv ('udemy_output_All_Finance__Accounting_p1_p626.csv')

finance_udemy.head()
```

```python
# general info on the dataset
finance_udemy.info()
```

## 2. Data cleaning

It is observed that some columns contain redundant values, hence these columns will be removed.

```python
#whether 'is_wishlisted' has true/false values
print ('is_wishlisted:', finance_udemy['is_wishlisted'].unique())

# whether title can be used as course identifier
print('title:', finance_udemy['title'].nunique()) #duplicated values, each course not identifiable by title
```

```python
#'is_wishlisted' only has 'false' values, so we can't use to compare other variables and it is redundant
finance_udemy.drop(['title','url','is_wishlisted','avg_rating','avg_rating_recent','created','discount_price__currency','discount_price__price_string','price_detail__price_string','price_detail__currency'],axis = 1,inplace=True)
finance_udemy.head()
```

```python
#ID as the index of the dataset
finance_udemy.set_index('id', inplace=True)
print(finance_udemy.index.name)
finance_udemy.head()
```

To avoid syntax error, certain columns will be renamed:

```python
#renaming certain columns
column_list = (list(finance_udemy.columns))
print(column_list)
column_list[7] = 'discount_amount'
column_list[8] = 'price_amount'
finance_udemy.columns = column_list
finance_udemy.columns
```

Next, we will check for any missing values:

```python
#check for missing values
finance_udemy.isna().sum()
```

By referring above, we understand that:
- The missing values in `discount_amount` is reasonable because not all courses offer discounts. 
- The missing values in `price_amount` refers to the free courses offered.

Hence, we will replace these missing values with $0$.

```python
finance_udemy.fillna(value = 0, inplace = True)

#check if there are any missing values
finance_udemy.isna().sum()
```

```python
#check the data types for each column
finance_udemy.info()
```

To provide a better context, we will include the price amount based on the average IDR-MYR exchange rate provided in year 2020.

```python
#include MYR rate for price amount
finance_udemy['priceamount_MYR'] = finance_udemy['price_amount'].apply(lambda x: x*0.057)

finance_udemy.head()
```

## 3. Exploratory analysis and visualization

To observe the trend of subscribing to accounting and finance related courses, we will group the data according to:
1. Years
2. Months

```python
#change date.dtype to date-time
finance_udemy['published_time'] = pd.to_datetime(finance_udemy['published_time'])

#grouping the rows according to the years
finance_udemy['published_year'] = finance_udemy['published_time'].dt.year

#grouping the rows according to months
finance_udemy['published_month'] = finance_udemy['published_time'].dt.month

year = finance_udemy.groupby('published_year').sum()
month = finance_udemy.groupby('published_month').sum()
```

```python
year
```

```python
month
```

As the sum of `num_subscribers` does not allow comparison between months (regardless of number of years), we will calculate the average number of subscribers for each month and create a new column `avg_subscribers`. 

```python
month['avg_subscribers']= month['num_subscribers'] / finance_udemy['published_year'].nunique()
month.head(12)
```

To provide a better data visualization based on months, we will create `month_dict` that includes the months of a year in words:

```python
month_dict = {
     1:'January',
     2:'February',
    3:'March',
     4:'April',
     5:'May',
     6:'June',
     7:'July',
     8:'August',
     9:'September',
     10:'October',
     11:'November',
     12:'December'
}

finance_udemy['month_in_words'] = finance_udemy['published_month'].map(month_dict)
```

```python
#rearranging columns
finance_udemy = finance_udemy[['is_paid','rating','num_reviews','num_subscribers','num_published_lectures','num_published_practice_tests','discount_amount',
                              'price_amount','priceamount_MYR','published_time','published_year','published_month','month_in_words']]

finance_udemy.head()
finance_udemy.describe()
```

## Descriptive Statistics
### Total courses offered from 2010 to 2020

```python
#total courses throughout 2010-2020
years = [year for year, finance_udemy in finance_udemy.groupby(['published_year'])]
total_courses = finance_udemy['published_year'].value_counts()
plt.bar(years, total_courses.sort_values())
plt.title('Overall Trend of Finance and Accounting Courses (2010-2020)')
plt.xlabel('Year')
plt.ylabel('Number of Courses')
plt.show()
```

There was a huge increase in the number of finance and accounting related courses offered from year 2014 to year 2018 but the rate of increase seems to slow down from 2018 to 2020.

### Total number of subscribers vs. Years

```python
#total subscribers vs years
plt.bar(years,finance_udemy.groupby(['published_year']).sum()['num_subscribers'], color = ['lightblue'])
plt.ylabel('Total Subscribers')
plt.xlabel('Year')
plt.title('Total Subscribers in Year 2010 - 2020')
plt.show()
```

The total number of subscribers for these courses increased rapidly from 2013 to 2017 and the highest number of subscribers was in year 2017. However, the graph shows a decrease in number after year 2017.


### Average number of subscribers vs. Months

```python
#average of subscribers in a year (to see if there's a certain pattern)
plt.bar (month_dict.values(), month['avg_subscribers'], color = ['lightgrey'])
plt.title('Average Subscribers in A Year')
plt.xlabel('Month')
plt.ylabel('Average Subscribers')
plt.xticks(rotation=30) 
plt.show()
```

The graph shows a high average number of subscribers in certain months (ie. March and April) while the lowest average number of subscribers is reported in June. This could provide insight for Udemy when considering marketing related-decisions.



Next, we will look at the correlation between continuous data.

```python
#correlational values
import seaborn as sb
fig=plt.subplots(1,1,figsize=(15,8))
sb.heatmap(finance_udemy.corr(), annot=True)
```

By looking at each correlational value for continuous data, we can see that:

We will investigate the top three relatively high correlational relationships by plotting scattergraph.

Even though the correlational relationship between number of published lectures and price amount is relatively weak (.27), it does show a slight pattern where higher number of published lectures would have higher price.

### Number of subscribers vs. Number of published lectures

The graph indicates a low correlational relationship between the number of subscribers and number of published lectures (.21), where courses with lectures as high as 110 has zero subscribers.

## 4. Real-life Implications
As we look into the general trend of accounting and finance courses offered in Udemy, we can have a rough idea of possible factors that influence potential subscribers to choose these courses. Below is an example of utilising two factors which may be considered by potential subscribers:

```python
#recommend courses based on starting month and budget
array = finance_udemy['month_in_words'].unique()
result = [(x) for x in finance_udemy['priceamount_MYR']]

def personal_preference (month,budget):
    if month in array or budget in result:
        if budget >= 0:
            recommended_course = finance_udemy[ (finance_udemy['month_in_words'] == month) & (finance_udemy['priceamount_MYR'] <= budget)]
            print (f'You prefer to start in ', month, ' and your budget is $', budget, sep = '')
            print ('')
            print (f'Here is a list of course IDs based on your preference:', recommended_course['priceamount_MYR'].index.tolist())
        else:
            print ('Please enter a correct amount.')
    else:
        print('Please enter your preferred starting month and budget again.')
```

```python
#kwargs
personal_preference(budget = 100, month = 'June')
```
